```python
Num_Cols = [
    'size_bytes',
    'price',
    'rating_count_tot',
    'rating_count_ver',
    'sup_devices.num',
    'ipadSc_urls.num',
    'lang.num',
    'cont_rating'
]

Cat_Cols = [
    'currency',
    'prime_genre'
]

Target_col = 'user_rating'

Data = Num_Cols + Cat_Cols + [Target_col]

AppStore = AppStore[Data]
AppStore['cont_rating'] = AppStore['cont_rating'].str.slice(0, -1).astype(int)
AppStore
```

**Задание** 

добавьте в датафрейм колонки для всех категориальных фичей и обновите список категориальных фичей

```python
AppStore = pd.get_dummies(AppStore)
AppStore
```

```python
pca = StandardScaler()
# Выход pca - numpy матрица, положим ее в новую переменную со всеми фичами
AppStore[AppStore.columns] = pca.fit_transform(AppStore[AppStore.columns])

AppStore
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(AppStore.drop(columns=['user_rating']), AppStore[Target_col], test_size=0.2)
```

**Задание**

Обучите несколько других моделей и получите лучшие скоры

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn = KNeighborsRegressor(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
print(knn.score(X_train, y_train))
print(knn.score(X_test, y_test))
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
reg = DecisionTreeRegressor(random_state=0)
reg.fit(X_train, y_train)
```

```python
print(reg.score(X_train, y_train))
print(reg.score(X_test, y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
for_reg = RandomForestRegressor(max_depth=5, random_state=0, n_estimators=50, n_jobs=-1)
for_reg.fit(X_train, y_train)
```

```python
print(for_reg.score(X_train, y_train))
print(for_reg.score(X_test, y_test))
```
